```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
from scipy.stats import sem
#from scipy import stats

import numpy as np

# Study data files
mouse_metadata = "Mouse_metadata.csv"
study_results = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

data_study = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
data_study
# Combine the data into a single dataset
```

```python
#Look across all previously generated figures and tables and write at least three observations or inferences that can be made from the data. Include these observations at the top of notebook.
#We can observe that the most promising treatment regimens for cancer are: Ramicane, Capomulin, Propriva, Ceftamin)
#We can see that there is a correlation between the weight of the mouse and the tumor volume of 0.82, when the mouse is
#treated with the medicine Capomulin. 
#Which means that the size of the tumor will be bigger if the weight of the mice is high.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
grouped_tumor_df=data_study.groupby(['Drug Regimen'])
grouped_tumor_df.count().head(12)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
volume_data= pd.DataFrame(grouped_tumor_df["Tumor Volume (mm3)"].mean())
volume_data.rename(columns={"Tumor Volume (mm3)":"Mean"},inplace=True)
volume_data_st=pd.DataFrame(grouped_tumor_df["Tumor Volume (mm3)"].std())
volume_data_st.rename(columns={"Tumor Volume (mm3)":"Standard deviation"},inplace=True)
volume_data_variance=pd.DataFrame(grouped_tumor_df["Tumor Volume (mm3)"].var())
volume_data_variance.rename(columns={"Tumor Volume (mm3)":"Variance"},inplace=True)
volume_data_median=pd.DataFrame(grouped_tumor_df["Tumor Volume (mm3)"].median())
volume_data_median.rename(columns={"Tumor Volume (mm3)":"Median"},inplace=True)

volume_data_sem=pd.DataFrame(grouped_tumor_df["Tumor Volume (mm3)"].sem())
volume_data_sem.rename(columns={"Tumor Volume (mm3)":"SEM"},inplace=True)

merge_df=pd.merge(volume_data, volume_data_st, on="Drug Regimen", how="outer")
merge_df_2=pd.merge(merge_df, volume_data_variance, on="Drug Regimen", how="outer")
merge_df_3=pd.merge(merge_df_2, volume_data_median, on="Drug Regimen", how="outer")
merge_df_4=pd.merge(merge_df_3, volume_data_sem, on="Drug Regimen", how="outer")
merge_df_4
```
